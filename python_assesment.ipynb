### Prime Number
- The function checks if a number is prime by testing whether any value from 2 up to the square root of the number divides it evenly.
- If no such divisor exists, the number is identified as prime


```python
def prime(n):
  if n<=1:
    return False
  for i in range(2,int(n*0.5)+1):
    if n%i==0:
      return False
    return True
print(prime(7))
```

### Printing Even Numbers from the List
- The function goes through each element in the list and checks whether it is an even number.
- Every even value is collected into a new list, which is then returned and printed.

```python
def lst(lst1):
  cleaned_lst=[]
  for i in range(len(lst1)):
    if lst1[i]%2==0:
      cleaned_lst.append(lst1[i])
  return cleaned_lst
   
lst1=[1,2,3,4,5,6]
print(lst(lst1))
```

### Tuple and Set Operations
- The function converts the list into a set to remove duplicates, then turns it back into a list.
- The cleaned list is reversed and returned as the final output.

```python
def lst(lst1):
  lst2=list(set(lst1))
  lst3=sorted(lst2,reverse=True)
  return lst3
print(lst([1,2,3,4,5,6,5,2,4,6,3,2,1,4]))
```

###  Dictionary where Keys are words and Values are the vowel counts
- The function loops through each name in the list and counts how many vowels appear in it by checking each character.
- For every name, the vowel count is stored and returned as a list of results matching each input word.

```python
def lst(lst1):
    count=[]
    vowels=['A','E','I','O','U','a','e','i','o','u']
    for name in lst1:
        vowel_count=0
        for char in name:
            if char in vowels:
                vowel_count+=1
        count.append(vowel_count)
    return list(zip(lst1,count))
print(lst(["Alice", "Bob", "Cathy", "David", "Ethan"]))

```

### Sum of Diagonal elements of a Matrix
- The program creates a NumPy matrix and extracts its main diagonal using np.diag(), which collects the elements where the row and column indices are the same.
- It then adds these diagonal values together and prints the total as the sum of the matrix’s diagonal elements.

```python
import numpy as np
matrix = np.arange(1,26).reshape(5,5)
diagonal_matrix=np.diag(matrix)
print(matrix)
print(diagonal_matrix.sum())
```

### Pandas DataFrame Filtering
- The program creates a DataFrame containing employee names, departments, and salaries, then calculates the 75th-percentile salary using quantile(0.75).
- It filters the DataFrame to show only those employees whose salaries are above this threshold, effectively listing top-earning employees.

```python
import pandas as pd
data = {
    "name": ["Alice", "Bob", "Cathy", "David", "Eva"],
    "department": ["HR", "IT", "Finance", "IT", "Marketing"],
    "salary": [45000, 60000, 55000, 70000, 65000]
}
df=pd.DataFrame(data)
threshold = df['salary'].quantile(0.75)
print(df[df['salary']>threshold])
```

### DataFrame Plotting
- The DataFrame is created with product categories and their sales values, and the total sales for each category are calculated.
- A bar chart is plotted using Matplotlib to visually compare the sales across categories, along with axis labels and a title.

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "category": ["Electronics", "Clothing", "Electronics", "Food", "Clothing"],
    "sales": [1200, 800, 1500, 600, 900]
}
df = pd.DataFrame(data)
category_sales = df.groupby("category")["sales"].sum()
print(category_sales)
category_sales.plot(kind="bar")

plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.title("Total Sales by Category")
plt.tight_layout()
plt.show()
```

### OOP – Class and Inheritance
- A base class is created to store common attributes like make and model, and a child class extends it by adding features specific to cars such as the number of doors.
- The child class uses inheritance to access the parent’s attributes, and an object is created to display complete information about a specific car.

```python
class vehicle:
    def __init__(self,make,model,):
        self.make = make
        self.model = model
class car(vehicle):
    def __init__(self,make,model,num_doors):
        super().__init__(make,model)
        self.num_doors = num_doors
    def display_info(self):
        print(f'{self.make} {self.model} has {self.num_doors} doors ')
car1 = car('Honda','Civic',4)
car1.display_info()
```

### OOP – Encapsulation and Methods
- The class stores account details and protects the balance through methods that safely handle deposits and withdrawals, ensuring only valid operations are allowed.
- After performing a withdrawal and a deposit, the updated balance is accessed through a method designed to return the current account value.

```python
class bank:
    def __init__(self,ac_name,balance=0.0):
        self.ac_name=ac_name
        self.__balance=balance
    def deposite(self,amount):
        if amount>0:
            self.__balance+=amount
            return True
        return False
    def withdraw(self,amount):
        if 0<= amount< self.__balance:
            self.__balance-=amount
            return True
        return False
    def get_balance(self):
        return self.__balance  

acc=bank('Vin',1500)
acc.withdraw(200)
acc.deposite(500)
print("Current Balance:", acc.get_balance())
print(acc)
```
